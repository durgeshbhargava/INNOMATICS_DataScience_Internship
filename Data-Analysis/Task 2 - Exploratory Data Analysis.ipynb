## Univariate analysis

```python
#plotting the feature values 
plt.figure(figsize=(20,25))
plotnumber=1
for column in df_num:
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        plt.plot(df_num[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
#plotting the feature values 
plt.figure(figsize=(20,25))
plotnumber=1
for column in df_num:
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        sns.histplot(df_num[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
#plotting the feature values 
plt.figure(figsize=(20,25))
plotnumber=1
for column in df_num:
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(df_num[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```
